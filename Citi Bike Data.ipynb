```python
import pandas as pd
```

```python
membership_dfs = []
membership_files = ["2013/Launch - Sep.csv", "2013/Oct - Dec.csv", 
         "2014/Jan - Mar.csv", "2014/Apr - Jun.csv", "2014/Jul - Sep.csv", "2014/Oct - Dec.csv",
         "2015/Jan - Mar.csv", "2015/Apr - Jun.csv", "2015/Jul - Sep.csv", "2015/Oct - Dec.csv",
         "2016/Jan - Mar.csv", "2016/Apr - Jun.csv", "2016/Jul - Sep.csv", "2016/Oct - Dec.csv",
         "2017/Jan - Mar.csv", "2017/Apr - Jun.csv", "2017/Jul - Sep.csv", "2017/Oct - Dec.csv",
         "2018/Jan - Mar.csv", "2018/Apr - Jun.csv", "2018/Jul - Sep.csv", "2018/Oct - Dec.csv",
         "2019/Jan - Mar.csv"]
for file_name in membership_files:
    membership_dfs.append(pd.read_csv(file_name))
membership_df = pd.concat(membership_dfs, axis=0)
```
